```python
import matplotlib.pyplot as plt
import numpy as np

from data_hoda import knn_classifier,load_hoda
```

```python
x_train, y_train, x_test, y_test = load_hoda()
```

```python
y=[]
x=list(range(2,16))
for k in x:
  y.append(knn_classifier(k,x_train,y_train,x_test,y_test))
```

```python
plt.plot(x,y)
plt.xlabel('k')
plt.ylabel('acc')
plt.title('optimal k')
plt.show()
```

I think "3" is the optimal k for knn-classification in hoda dataset
