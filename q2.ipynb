```python
max_acc=0
max_train_size=1000

for tsize in range(1000,5800,100):
  x_train,y_train,x_test,y_test=load_hoda(size=20,training_sample_size=tsize)
  acc=knn_classifier(3,x_train,y_train,x_test,y_test)
  if acc>max_acc:
    max_acc=acc
    max_train_size=tsize
print("max acc= ",max_acc,"max train size= ",max_train_size)
```
